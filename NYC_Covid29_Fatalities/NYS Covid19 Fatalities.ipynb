```python
import pandas as pd
```

```python
# File to Load (Remember to change the path if needed.)
NYS_COVID19_fatalities_data = "New_York_State_Statewide_COVID19_Fatalities_by_County.csv"
```

```python
# Read CSV into a Pandas DataFrame
NYS_covid19_fatalities_data_df = pd.read_csv(NYS_COVID19_fatalities_data)

NYS_covid19_fatalities_data_df
```

```python
NYC_covid19_fatalities_data_df = NYS_covid19_fatalities_data_df[NYS_covid19_fatalities_data_df["County"].isin(['Bronx', 'Manhattan', 'Kings', 'Queens', 'Richmond'])]

NYC_covid19_fatalities_data_df.head(15)
```

```python
NYC_covid19_fatalities_data_df.to_csv('NYC_COVID19_FatalitiesTable',index=False)
```
