```python
import pandas as pd
```

```python
# File to Load (Remember to change the path if needed.)
NYS_COVID19_DailyHospitalizationSummary_data = "New_York_Forward_COVID-19_Daily_Hospitalization_Summary_by_Region.csv"

NYS_COVID19_DailyHospitalizationSummary_data
```

```python
# Read CSV into a Pandas DataFrame
NYS_covid19_DailyHospitalizationSummary_data_df = pd.read_csv(NYS_COVID19_DailyHospitalizationSummary_data)

NYS_covid19_DailyHospitalizationSummary_data_df
```

```python
NYC_covid19_DailyHospitalization_data_df = NYS_covid19_DailyHospitalizationSummary_data_df[NYS_covid19_DailyHospitalizationSummary_data_df["Region"].isin(['NEW YORK CITY'])]

NYC_covid19_DailyHospitalization_data_df.head(15)
```

```python
NYC_covid19_DailyHospitalization_data_df.to_csv('NYC_COVID19_DailyHospitalizationTable',index=False)
```
